```python
def euclidean_distance(pt1, pt2):
    distance = np.sqrt(np.sum(pt1 - pt2)**2)
    return distance

# Testing the function
a = np.array([3, 6])
b = np.array([4, 2])
print(euclidean_distance(a, b))
```

```python
def KNN(X_train, X_test, Y_train, Y_test, k_val):
    y_hat = []
    
    
    for test_pt in X_test.to_numpy():
        distances = []
        for i in range(len(X_train)):
            point1 = np.array(X_train.iloc[i])
            distances.append(euclidean_distance(point1, test_pt))
        
        distance_data = pd.DataFrame(data = distances, columns=["distance"], index = Y_train.index)
        
        # Sort the data into K closest neighbors
        k_neighbors_list = distance_data.sort_values(by=["distance"], axis=0)[:k_val]
        
        # Find the labels of the neighbors
        labels = Y_train.loc[k_neighbors_list.index]
        
        # Mode tells us the most occurences.
        voting = mode(labels).mode[0]
        y_hat.append(voting)
        
        print(y_hat)
    return y_hat
```

```python
y_hat_test = KNN(X_train2, X_test2, Y_train2, Y_test2, k_val=5)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy of model: ", accuracy_score(Y_test2, y_hat_test))
```
